* Choose another type of plot that interests you from the [pandas.DataFrame.plot documentation](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.html#) [look at the 'kind' parameter] and make a new plot of your dataset values using the plot type

```python
# Define function unique that takes in any list
# outputs unique values in this list 
def unique(list):
    unique_list = []
    for i in list:
        if i not in unique_list:
            unique_list.append(i)
    return unique_list

# use unique to find an unique list of the states in this dataset
state_list=unique(list(df["State"].values))

# Each subplot below is the Job Creation Rate versus Year 
# for Each American State from 1980 ~ 2019 

for state in state_list:
    new_df = df[df["State"] == state]
    new_df.plot(kind="scatter",x="Year",y="Data.Job Creation.Rate/Births",
                title=state,
                ylabel="Job Creation Rate")
    
# The overall trend for all states is decreasing as time goes on. 
# An exception could be Montana, which has been seeing an increase for the past few years. 
```

```python
# Below is what I did last week, which finds the number of firms in California
# for each year from 1978 to 2018 

def find_num_firms(file,state,year,column):
    # take only rows in the wanted year 
    new = file[file["Year"] == year]
    
    # result is the value of the column for that specific state and year wanted
    result = new[new["State"] == state][column]
    
    return result 

for year in list(range(1978,2019)):
    print("State: California |","Year:",year, "| Column:",find_num_firms(file=df,state="California",year=year,column = "Data.Number of Firms").values[0],sep = " ")
```

```python
# Now, I will extend my analysis by creating a scatterplot 
# that shows the trend of number of firms in California for 1978 ~ 2018 

cali_df = df[df["State"] == "California"]
cali_df.plot(kind="scatter",x="Year",y="Data.Number of Firms",
            title="Number of firms in California vs. Year from 1978 to 2018",
            ylabel="Number of Firms")

# We could see that the general trend for firms in California is increasing. 
# From its scatterplot of job creation rate I did in cell 85, 
# we see that the job creation rate has been decreasing in California overtime, 
# in contrast to the number of firms here. 
```

```python
# I furthered my analysis into job creations in California. 
# From the two graphs below,
# we could see that, although the number of job creation in California has been increasing,
# which is in coherence with the number of firms in California over 1978 ~ 2018 trend,
# the Net rate of which the jobs were created during this period, has a mildly decreasing trend,
# corresponding to the decreasing job birth rate. 

# Thus, we could possibly conclude that, though the number of firms in California is increasing overtime,
# the number of net jobs they create has been decreasing. 
# We could even postulate that, since the number of jobs are decreasing overtime,
# profits might be as well decreasing, though could be masked by the inflation rate overtime. 

cali_df.plot(kind="scatter",x="Year",y="Data.Calculated.Net Job Creation",
            title="Calculated Net Job Creation in California vs. Year from 1978 to 2018",
            ylabel="Calculated Net Job Creation")


cali_df.plot(kind="scatter",x="Year",y="Data.Calculated.Net Job Creation Rate",
            title="Calculated Net Job Creation Rate in California vs. Year from 1978 to 2018",
            ylabel="Calculated Net Job Creation Rate")
```
